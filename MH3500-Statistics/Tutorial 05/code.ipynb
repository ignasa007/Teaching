```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns

mlp.rcParams['font.family'] = 'serif'
mlp.rcParams['font.size'] = 16
```

```python
# Problem 5.3(b)

def theta_1(sample):
    return (-1/2) * np.mean(sample)
def theta_2(sample):
    return (1/6) * (7-np.sum(sample[:3]))
def theta_3(sample):
    return (1/6) * (7-3*np.mean(sample))
def theta_4(sample):
    return (1/16) * (17-3*np.mean(np.square(sample)))

theta = 0.5
n = 100
n_samples = 10_000

theta_hat_1, theta_hat_2, theta_hat_3, theta_hat_4 = list(), list(), list(), list()
for _ in range(n_samples):
    sample = np.random.choice(a=(0,1,2,3), p=((2/3)*theta,(1/3)*theta,(2/3)*(1-theta),(1/3)*(1-theta)), size=n)
    theta_hat_1.append(theta_1(sample)); theta_hat_2.append(theta_2(sample)); theta_hat_3.append(theta_3(sample)); theta_hat_4.append(theta_4(sample))

plt.figure(figsize=(8,6))
plt.hist(theta_hat_1, bins=20, histtype='step', label=r'$\hat{\theta}_1$')
plt.hist(theta_hat_2, bins=20, histtype='step', label=r'$\hat{\theta}_2$') 
plt.hist(theta_hat_3, bins=20, histtype='step', label=r'$\hat{\theta}_3$') 
plt.hist(theta_hat_4, bins=20, histtype='step', label=r'$\hat{\theta}_4$')
plt.xlabel('Estimates', labelpad=12)
plt.ylabel('Frequency', labelpad=12)
plt.legend()
plt.grid()
plt.show()
```
